```python
#Importing the required modules
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

geolocator = Nominatim(user_agent="my_request")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
```

```python
address_df = pd.DataFrame({'add': address})
address_df['location'] = address_df['add'].apply(geocode)
address_df['Lat'] = address_df['location'].apply(lambda x: x.latitude if x else None)
address_df['Lon'] = address_df['location'].apply(lambda x: x.longitude if x else None)
 
address_df
```

```python
from datetime import datetime         #función que recibe una dirección y hace la llamada la libreria geocoder para recibir un json con datos de esta dirección (lat, long, código postal,...)

def function_address(address):
    
    #start = datetime.now()
    try:
        response = geocoder.osm(address + ", Madrid").json
        
    except: response = None
        
    #print(datetime.now() - start)
    return response
function_address("Calle de Godella, 64")
```

```python
new_address = pd.DataFrame({'add': df2.raw_address})    #recibe todas las filas de la columna donde están las direcciones y devuelve localización.
new_address['location'] = new_address['add'].apply(function_address)

 
new_address 
```

```python
def coordinate(x,typ):    #Del json recibido separo longitud y latitud.
    try:
        return x[typ]
        
    except: return None
```

```python
new_address["latitud"] = new_address.location.apply(coordinate, typ = "lat")
new_address["longitude"] = new_address.location.apply(coordinate, typ = "lng")
new_address
```
